```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import quantile_transform
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import OneHotEncoder  ##. better to use dummy from pandas 
from sklearn.preprocessing import PowerTransformer
from scipy.stats import boxcox
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from scipy.stats import boxcox
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
pd.options.display.max_rows = 50
```

```python
df_customer = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
df_customer
```

```python
numerical=df_customer.select_dtypes(int)
numerical=numerical.drop("Unnamed: 0", axis=1)
numerical
```

### X - y Split

```python
X=numerical.drop('total_claim_amount', axis=1)
y=df_customer.total_claim_amount
```

### Train 70% and Test 30%

```python
# Train-test split.
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.30,random_state=123)
```

```python
# Standardize the data (after the data split).

std_scaler=StandardScaler().fit(X_train)  
X_train_scaled=std_scaler.transform(X_train)

X_train_scaled
```

```python
X_train.shape
```

```python
X_test_scaled=std_scaler.transform(X_test)
X_test_scaled
```

```python
X.columns
```

```python
# Apply linear regression.
model=LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Model Interpretation.
model.coef_
```

```python
model.intercept_
```

```python
model.score(X_test_scaled,y_test)
```

### Predicting

```python
# Prediction
y_pred=model.predict(X_test_scaled)
y_pred
```

```python
y_test
```

```python
result=pd.DataFrame({"y_test":y_test,"y_pred":y_pred})
result
```

```python
# scatter plot y_pred vs y
fig, ax = plt.subplots(1,3,figsize=(14,4))

ax[0].plot(y_pred, y_test, 'o')
ax[0].set_xlabel("y_test")
ax[0].set_ylabel("y_pred")
ax[0].set_title("Test Set -Predicted vs real")

# histogram of the residuals between y - y_pred. 
# Homoscdasticity
ax[1].hist(y_test - y_pred)
ax[1].set_xlabel("Test y-y_pred")
ax[1].set_title("Test Set Residual histogram")

ax[2].plot(y_pred,y_test - y_pred,"o")
ax[2].set_xlabel("predited")
ax[2].set_ylabel("residuals")
ax[2].set_title("Residuals by Predicted")
ax[2].plot(y_pred,np.zeros(len(y_pred)),linestyle='dashed')
```

## Evaluation of the Model

```python
# MSE - Mean Square Error
mse(y_test,y_pred)
```

```python
# RMSE - Root Mean Square Error
mse(y_test, y_pred, squared=False)
```

```python
# MAE - Mean Absolute Error
mae (y_test, y_pred)
```

```python
# R2 - percentage of how good the model is
R2=r2_score(y_test,y_pred)
R2
```

```python
# Adjusted R2
Adj_R2= 1 - (1-R2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
Adj_R2
```

```python
# Feature Importance
features_importances = pd.DataFrame(data={'Attribute': X_train.columns,'Importance': abs(model.coef_)})
features_importances = features_importances.sort_values(by='Importance', ascending=False)
features_importances
```

```python
plt.bar(x=features_importances['Attribute'], height=features_importances['Importance'])
plt.title('Feature importances based on coefficients', size=15)
plt.xticks(rotation='vertical')
plt.show()
```

### Checking the categorical columns

#### Dropping the redundant columns

```python
categoricals=df_customer.select_dtypes(object)
categoricals=categoricals.drop(["effective_to_date"], axis=1)
categoricals
```

```python
categoricals=categoricals.drop(["policy"], axis=1)
categoricals
```

```python
# binning
categoricals["vehicle_class"]= categoricals["vehicle_class"].replace({"luxury car":"luxury", "luxury suv":"luxury"})
categoricals["vehicle_class"].value_counts()
```

### Get Dummies for the categorical columns

```python
categorical_dummies=pd.get_dummies(categoricals, drop_first=True)
categorical_dummies
```

## Concatenate numerical and categorical together

```python
def load_new_data():
    return pd.concat([numerical,categorical_dummies], axis=1)
```

```python
whole_df=load_new_data()
whole_df
```
