```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_customer = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
df_customer
```

```python
numerical=df_customer.select_dtypes(int)
numerical
```

```python
categoricals=df_customer.select_dtypes(object)
categoricals
```

```python
#Use seaborn library to construct distribution plots for the numerical variables
sns.displot((numerical), bins=10)
```

```python
#construct histogram
numerical.hist(figsize=(15,15))
```

```python
#Use Matplotlib to construct histogram

# create the figure and axes
fig,axes = plt.subplots(3,3,figsize = (15,15))

# add an overall title for the figure
fig.suptitle("Using Matplotlib to build a histogram", fontsize=18)

# create first plot (left), adding a title
axes[0,0].hist(numerical["income"], color='blue')
axes[0,0].set_title('income', fontsize=10)

# create second plot (centre), adding a title
axes[0,1].hist(numerical["number_of_policies"], color="purple")
axes[0,1].set_title('number of policies', fontsize=10)

# create third plot (right), adding a title
axes[0,2].hist(numerical["total_claim_amount"], color="green")
axes[0,2].set_title("total_claim_amount", fontsize=10)

axes[1,0].hist(numerical["customer_lifetime_value"], color="turquoise")
axes[1,0].set_title("customer_lifetime_value", fontsize=10)

axes[1,1].hist(numerical["number_of_open_complaints"], color="orange")
axes[1,1].set_title("number_of_open_complains", fontsize=10)

axes[1,2].hist(numerical["months_since_last_claim"], color="pink")
axes[1,2].set_title("months_since_last_claim", fontsize=10)

axes[2,0].hist(numerical["monthly_premium_auto"], color="black")
axes[2,0].set_title("monthly_premium_auto", fontsize=10)

axes[2,1].hist(numerical["months_since_policy_inception"], color="violet")
axes[2,1].set_title("months_since_policy_inception", fontsize=10)
```

```python
#Do the distributions for different numerical variables look like a normal distribution 
sns.displot((numerical["months_since_policy_inception"]), kde=True)
```

```python
numerical.corr()
```

```python
#For the numerical variables, check the multicollinearity between the features. 
#Please note that we will use the column `total_claim_amount` later as the target variable.
num_corr = numerical.corr()
mask = np.zeros_like(num_corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(10, 7))
    ax = sns.heatmap(num_corr, mask=mask,cmap='coolwarm', vmin=-1,vmax=1,annot=True, square=True)
```

```python
#Show a plot of the total number of responses.
sns.displot((categoricals["response"]), bins=5, color="purple")
```

```python
#Show a plot of the response rate by the sales channel.
plt.title('Response Rate based on Sales Channel', size='15', fontweight='bold')
sns.countplot(x ='sales_channel', hue ='response',data=categoricals)
```

```python
#Show a plot of the response rate by the total claim amount.
plt.title("Response rate based on claim amount", size="15", fontweight="bold")
sns.countplot(x="total_claim_amount", hue="response", data=df_customer)
```

```python
my_bins=[0, 500, 1000, 1500, 2000, 25000]
my_labels=["very low", "low", "medium", "high", "very high"]
df_customer["range_total_claim"]=pd.cut(df_customer["total_claim_amount"],bins=my_bins, labels=my_labels)
plt.title("Response rate based on claim amount", size="15", fontweight="bold")
sns.countplot(x="range_total_claim", hue="response", data=df_customer)
```

```python
#Show a plot of the response rate by income.
bins_income=[0, 20000, 40000, 60000, 80000, 100000]
labels_income=["very low","low", "medium", "high", "very high" ]
df_customer["range_income"]=pd.cut(df_customer["income"],bins=bins_income, labels=labels_income)
plt.title("Response rate based on Income", size="15", fontweight="bold")
sns.countplot(x="range_income", hue="response", data=df_customer)
```
