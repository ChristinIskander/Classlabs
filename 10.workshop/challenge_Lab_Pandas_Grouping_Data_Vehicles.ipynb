<b>show the average CO2_Emission_Grams/Km  by Brand

```python
### your code us here
cars.groupby('Make')["CO2 Emission Grams/Mile"].mean()
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
### your code us here
cars.groupby('Make')["CO2 Emission Grams/Mile"].mean().reset_index().sort_values(["CO2 Emission Grams/Mile"], ascending=False)
```

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
cars['Year'].unique()
```

```python
## your code here
labels=["1983-1992", "1992-2000", "2000-2008", "2008-2017"]
cars["Year Ranges"]=pd.cut(cars["Year"],bins=4, labels=labels)
cars
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
cars[["City MPG"]] = cars[["City MPG"]]*1.60934 / 3.78541
cars=cars.rename(columns={"City MPG":"City Km/Liter"})
```

```python
### your code is here
cars.groupby("Year Ranges")["City Km/Liter"].mean().reset_index().sort_values(["City Km/Liter"], ascending=False)
```

Which brands are more environment friendly?

```python
### your code is here
cars.groupby('Make')["CO2 Emission Grams/Mile"].mean().reset_index().sort_values(["CO2 Emission Grams/Mile"], ascending=True)
```

Does the drivetrain affect fuel consumption?

```python
## Your Code here
cars.groupby("Drivetrain")["Fuel Barrels/Year"].mean().reset_index().sort_values(["Fuel Barrels/Year"], ascending=False)
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
## Your Code here
cars.groupby("Transmission")["Fuel Barrels/Year"].mean().reset_index().sort_values(["Fuel Barrels/Year"], ascending=False)
```
