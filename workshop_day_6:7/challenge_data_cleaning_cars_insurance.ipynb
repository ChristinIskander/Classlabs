```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox

pd.options.display.max_rows = 50
```

```python
file1=pd.read_csv("file1.csv")
file1
```

```python
file2=pd.read_csv("file2.csv")
file2
```

```python
file3=pd.read_csv("file3.csv")
file3.rename(columns={'State':'ST','Gender':'GENDER'}, inplace=True )
file3
```

```python
def load_original_data():
    return pd.concat([file1,file2,file3], axis=0)
```

```python
cars_insurance=load_original_data()
cars_insurance
```

```python
def lower_case_column_names(cars_insurance):
    cars_insurance.columns=[i.lower() for i in cars_insurance.columns]
    return cars_insurance
```

```python
cars_insurance=lower_case_column_names(cars_insurance)
cars_insurance
```

```python
cars_insurance["customer"] = cars_insurance.drop(columns=["customer"], inplace=True, axis=1)
cars_insurance
```

```python
cars_insurance = cars_insurance[["st","gender","education", "policy type","vehicle class","customer lifetime value","income","monthly premium auto","number of open complaints", "total claim amount"]]
cars_insurance
```

```python
cars_insurance.info()
```

```python
#remove the percentage from the customer lifetime value and truncate it to an integer value.
cars_insurance['customer lifetime value'] = cars_insurance['customer lifetime value'].astype("string")
```

```python
cars_insurance['customer lifetime value'] = cars_insurance['customer lifetime value'].str.rstrip("%")
cars_insurance
```

```python
cars_insurance["customer lifetime value"] = pd.to_numeric(cars_insurance["customer lifetime value"])
cars_insurance.info()
```

```python
len(cars_insurance[cars_insurance['customer lifetime value'].isna()==True])
```

```python
#Filter the data in state and gender column
cars_insurance['gender'].value_counts(dropna=False)
```

```python
cars_insurance["gender"]=cars_insurance['gender'].replace({"Male":"M", "Female":"F", "female":"F","Femal":"F"})
cars_insurance['gender'].value_counts()
```

```python
cars_insurance["st"].value_counts()
```

```python
cars_insurance["st"]=cars_insurance["st"].replace({"Cali":"California", "AZ":"Arizona", "WA":"Washington"})
cars_insurance["st"].value_counts()
```

```python
#Removing duplicates
cars_insurance.drop_duplicates()
cars_insurance.info()
```

```python
#clean the number of open complaints and extract the middle number
cars_insurance["number of open complaints"].unique()
```

```python
cars_insurance["number of open complaintsaints"]=cars_insurance["number of open complaints"].astype("string")
```

```python
cars_insurance["number of open complaints"]=cars_insurance["number of open complaints"].str[:3].str[-1]
cars_insurance
```

```python
cars_insurance["number of open complaints"]=pd.to_numeric(cars_insurance["number of open complaints"])
cars_insurance.info()
```

```python
#Replace missing values with means of the column (for numerical columns)
income_mean=cars_insurance["income"][cars_insurance["income"]!=0].mean()
total_claim_amount_mean=cars_insurance["total claim amount"][cars_insurance["total claim amount"]!=0].mean()
monthly_premium_auto_mean=cars_insurance["monthly premium auto"][cars_insurance["monthly premium auto"]!=0].mean()
```

```python
cars_insurance["income"].fillna(income_mean, inplace=True)
cars_insurance["total claim amount"].fillna(total_claim_amount_mean, inplace=True)
cars_insurance["monthly premium auto"].fillna(monthly_premium_auto_mean, inplace=True)
```

```python
cars_insurance["income"].isna().sum()
cars_insurance["total claim amount"].isna().sum()
cars_insurance["monthly premium auto"].isna().sum()
```

```python
#Replacing null values
cars_insurance["income"].replace(0,income_mean)
cars_insurance["total claim amount"].replace(0,total_claim_amount_mean)
cars_insurance["monthly premium auto"].replace(0,monthly_premium_auto_mean)
```

```python
#replace column "State" to different zones.
cars_insurance["st"]=cars_insurance["st"].replace({"California":"West Region", "Arizona":"Central", "Washington":"East", "Oregon":"North West", "Nevada":"Central"})
cars_insurance["st"].value_counts()
```

```python
#Use string functions to standardize the text data (lower case)
cars_insurance = cars_insurance.applymap(lambda x: x.lower() if type (x)==str else x)
cars_insurance
```
